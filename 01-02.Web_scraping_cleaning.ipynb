## Web Scraping

- To get artist, timetable and stages 

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://clashfinder.com/m/p202x1/?user=0vlvek.ad"
```

<b> Download HTML with a get request

```python
response = requests.get(url)
response.status_code # 200 status code means OK!
```

### Creating the soup

```python
# parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
# check that the html code looks like it should
soup
```

```python
# Extracting all the info we need per each day
```

```python
all = soup.find_all("div",{"class":"day"})
```

```python
all[0]  # Wednesday 1st June
all[1]  #Thursday 2nd June
```

- On the line-up on the website includes concerts from 1st of June to Sunday 12nd of June, but the "official" festival goes from 2nd to 4th of June and 9th to 11th of June, the first two weekends of the month.
- That's why I just want to scrape the concert and the hours of these days.

```python
def scrape_to_df(number_of_day):
    day = all[number_of_day] #2 june

    names=[]
    artist = day.find_all("span",{"class":"actNm"})
    for row in artist:          # Print all occurrences
        n = row.get_text()
        names.append(n)
    hour=[]
    h_june = day.find_all("span",{"class":"actTime"})
    for row in h_june:          # Print all occurrences
        h = row.get_text()
        hour.append(h)

    d = {'Artist':names,'Hour':hour}
    df = pd.DataFrame(d)
    return df
```

```python
thursday_2 = scrape_to_df(1)
friday_3= scrape_to_df(2)
saturday_4 = scrape_to_df(3)
thursday_9 = scrape_to_df(8)
friday_10 = scrape_to_df(9)
saturday_11 = scrape_to_df(10)
```
